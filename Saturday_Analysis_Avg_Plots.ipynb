```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import pandas as pd
import numpy as np
import os
import requests
import json
from pprint import pprint
sns.set()
```

```python
means_df = pd.read_csv("Resources/saturday_ouput.csv")
means_df
```

```python
#Set context, increase font size
sns.set_context("poster", font_scale=1.5)
#Create a figure
plt.figure(figsize=(15,20))
#Define the axis object
ax = sns.barplot(x='Average Ratings (1-5)', y='Stadium Name', data=means_df, palette="OrRd")
#set paramters
ax.set(xlabel='Avg Rating (1-5)', ylabel='Stadium Name', title= "AVG Restaurant Rating in 2mi Radius")
plt.savefig()
```
