```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
import os
import requests
import json
from pprint import pprint
sns.set()
```

```python
sat_means_df = pd.read_csv("Resources/saturday1000_ouput.csv")
```

```python
mon_means_df = pd.read_csv("Resources/Monday1000_ouput.csv")
```

```python
combined_data = pd.merge(sat_means_df, mon_means_df, on="Stadium Name")
```

```python
clean_data = combined_data.drop([combined_data.index[18] , combined_data.index[19] , combined_data.index[20], combined_data.index[26] , combined_data.index[27] , combined_data.index[28]])
```

```python
pos = list(range(len(clean_data['Average Ratings (1-5)_x']))) 
width = 0.25 
fig, ax = plt.subplots(figsize=(50,20))

plt.bar(pos, 
        #using df['pre_score'] data,
        clean_data['Average Ratings (1-5)_x'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='r', 
        # with label the first value in first_name
        label=clean_data['Stadium Name'][0])

plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        clean_data['Average Ratings (1-5)_y'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='b', 
        # with label the second value in first_name
        label=clean_data['Stadium Name'][1])

ax.set_ylabel('Avg Rating')
ax.set_title('Average Rating')
ax.set_xticks([p + 0.5 * width for p in pos])
ax.set_xticklabels(clean_data['Stadium Name'], rotation="vertical")
plt.xlim(min(pos)-width, max(pos)+width*1)
plt.ylim([0, max(clean_data['Average Ratings (1-5)_x'] + clean_data['Average Ratings (1-5)_y'])])
plt.legend(['Average Rating Saturday', 'Average Rating Monday'], loc='upper left')
plt.tight_layout()
plt.grid()
plt.savefig("ChrisBranch_Images/AverageRating.png")
plt.show()
```

```python
#Set context, increase font size
sns.set_context("poster", font_scale=1.5)
#Create a figure
plt.figure(figsize=(15,20))
#Define the axis object
ax = sns.barplot(x='Average Price (1-4)_x', y='Stadium Name', data=clean_data, color="midnightblue")
#set paramters
ax.set(xlabel='Avg Price', ylabel='Stadium Name', title= "Avg Restaurant Price")

plt.savefig("ChrisBranch_Images/AveragePrice.png")
           
```
