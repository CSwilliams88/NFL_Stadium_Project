```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import pandas as pd
import numpy as np
import os
import requests
import json
from pprint import pprint
sns.set()
```

```python
teams_df = pd.read_csv("Resources/Monday1000_ouput.csv")
teams_df.head()
```

```python
sea_team = teams_df.loc[teams_df['NFL Team'] == 'Seattle Seahawks']
gb_team = teams_df.loc[teams_df['NFL Team'] == 'Green Bay Packers']
pit_team = teams_df.loc[teams_df['NFL Team'] == 'Pittsburgh Steelers']
ne_team = teams_df.loc[teams_df['NFL Team'] == 'New England Patriots']
dal_team = teams_df.loc[teams_df['NFL Team'] == 'Dallas Cowboys']

pop_team_df = pd.concat([sea_team, gb_team, pit_team, ne_team, dal_team])
pop_team_df

ny_city = teams_df.loc[teams_df['NFL Team'] == 'New York Giants']
chi_city = teams_df.loc[teams_df['NFL Team'] == 'Chicago Bears']
lv_city = teams_df.loc[teams_df['NFL Team'] == 'Las Vegas Raiders']
sf_city = teams_df.loc[teams_df['NFL Team'] == 'San Francisco 49ers']

pop_city_df = pd.concat([ny_city, chi_city, lv_city, sea_team, sf_city])
pop_city_df

jack = teams_df.loc[teams_df['NFL Team'] == 'Jacksonville Jaguars']
tb = teams_df.loc[teams_df['NFL Team'] == 'Tampa Bay Buccaners']
ten = teams_df.loc[teams_df['NFL Team'] == 'Tennessee Titans']
cin = teams_df.loc[teams_df['NFL Team'] == 'Cincinnati Bengals']
buf = teams_df.loc[teams_df['NFL Team'] == 'Buffalo Bills']


not_pop_team = pd.concat([jack, tb, ten, cin, buf])
not_pop_team

comb_team = pd.concat([pop_team_df, pop_city_df, not_pop_team])
comb_team
```
